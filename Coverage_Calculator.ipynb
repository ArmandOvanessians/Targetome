```python
import numpy as np
import ast
import pandas as pd

def get_chain_coverage(pdb_id, chains):
    # Turn chains into a list of strings
    chains = eval(chains)
    coverage = 0 
    for chain in chains:
        atom_file_path = f"pdb_files/{pdb_id}.apdb"
        chain_id = chain
        residue_numbers = set()  # Initialize a set to store unique residue numbers

        try: 
            with open(atom_file_path, 'r') as atom_file:
                for line in atom_file:
                    if line.startswith('ATOM'):
                        current_chain_id = line[21]
                        if current_chain_id == chain_id:
                            residue_number = int(line[22:26].strip())
                            residue_numbers.add(residue_number)  # Add residue number to the set

        except: 
            return 0

        if not residue_numbers:  # Check if the set is empty
            return 0

        coverage += len(residue_numbers) # Increment coverage by the number of unique residue numbers in the set
    #print(f"PDB_ID : {residue_numbers}")
    return coverage
```

```python
path = "Uniport_human_AF2_Ranked_Chains_Filtered_missing_confidence.xlsx"
df_full = pd.read_excel(path)
df_full
```

```python
df_full["Coverage"] = df_full.apply(lambda row: get_chain_coverage(row["BestRanked"], row["Chain_id"]), axis=1)
```

```python
df_full["Percent Covered"] = df_full.apply(lambda row: row["Coverage"] / row["Length"] * 100, axis=1)
```

### There are 472 out of 7603 PDB files where coverage is > 100 % this could be due to 
 - negative residues present in the pdb file
 - discreptencies in PDB file chain naming (authors vs uniprot) 
 - Post-translational modifcation can introduce additional atoms or residues in the protein structure 
 - different isoforms with varying lenghts may exist
