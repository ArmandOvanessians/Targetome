```python
def get_residue_confidence_scores(pdb, residue_numbers):
    try:
        residue_numbers = ast.literal_eval(residue_numbers)
    except:
        return pd.Series([np.nan, np.nan, np.nan, np.nan])
    #print(f"FILE : {pdb}")
    # print(f"Chain_id type : {type(chain_id)} Residue_number type: {residue_numbers}")

    #AlphaFoldProt/AF-A0A087WT03-F1.pdb
    pdb_file_path = f"AlphaFoldProt/AF-{pdb}-F1.pdb"
    with open(pdb_file_path, 'r') as file:
        lines = file.readlines()
    # initialize missing scores
    scores = {}
    # initialize present scores 
    scores_present = {}
    #print(residue_numbers)
    #print(type(residue_numbers))
    tot_res_missing = 0 
    tot_scores_missing = 0
    tot_res_present = 0  
    tot_scores_present = 0 
    for i in range(len(residue_numbers)):
        #ID = chain_id[i]
        residues = residue_numbers[i]
    
        #print(f"Residues {residue_numbers[i]}")
        #print(residues)
        #print(residues)
        #scores = {}
            
        for line in lines:
            if line.startswith('ATOM'):
                atom_name = line[12:16].strip()
                try: 
                    residue_number = int(line[22:26].strip())
                except:
                    continue
                # if residue is one of the residue numbers missing                                                            
                if residue_number in residues:
                    #print("hi")
                    
                    try:
                        score = float(line[60:66].strip())
                        
                    except:
                        continue
                    tot_scores_missing += score
                    tot_res_missing += 1
                    index = residue_number
                    scores[index] = score
                # Else residue is present in the structure
                else:
                    score_present = float(line[60:66].strip())
                    tot_scores_present += score_present
                    tot_res_present += 1
                    index = residue_number
                    scores_present[index] = score_present
                    
                    
    
    if tot_res_missing == 0:
        avg_missing = np.nan
    else:
        avg_missing = tot_scores_missing/tot_res_missing
    
    if tot_res_present == 0:
        avg_present = np.nan
    else:
        avg_present = tot_scores_present/tot_res_present
        
                                                                                            
    return pd.Series([scores, avg_missing, scores_present, avg_present])
```

```python
def get_residue_confidence_scores(ID, chain_ids, missing_residues):
    pdb_file_path = f"AlphaFoldProt/AF-{ID}-F1.pdb"
    with open(pdb_file_path, 'r') as file:
        lines = file.readlines()

    scores = {}

    for chain_id, residues in zip(chain_ids, missing_residues):
        for line in lines:
            if line.startswith('ATOM'):
                atom_name = line[12:16].strip()
                chain = line[21]
                residue_number = int(line[22:26].strip())
                
                if chain == chain_id and atom_name == 'CA' and residue_number in residues:
                    score = float(line[60:66].strip())
                    scores[(chain_id, residue_number)] = score

    return scores
```
