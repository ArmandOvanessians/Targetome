```python
import pandas as pd
import functions
from functions import UniprotDataFetcher
from functions import PDB_analyzer
import requests
import ast
from Bio.PDB import PDBParser
import numpy as np
import pandas as pd
import warnings
from Bio import BiopythonParserWarning
warnings.simplefilter('ignore', BiopythonParserWarning)
from Bio import BiopythonExperimentalWarning
warnings.filterwarnings('ignore')
```

```python
def find_missing_residues(pdb, chain_id):
    tot =[]
    for chain in chain_id:
        # get protein path
        pdb_file = f"pdb_files/{pdb}.apdb"
        # Create a PDB parse object
        parser = PDBParser()
        with warnings.catch_warnings():
            warnings.simplefilter('ignore', BiopythonExperimentalWarning)
            structure = parser.get_structure('protein', pdb_file)

        # Get the desired chain
        model = structure[0]  # Assuming single model
        chain = model[chain]

        # Extract the residue numbers present in the chain
        present_residues = {residue.get_id()[1] for residue in chain.get_residues()}

        # Find missing residues
        max_residue_num = max(present_residues)
        missing_residues = [residue_num for residue_num in range(1, max_residue_num + 1) if residue_num not in present_residues]
        tot.append(missing_residues)

    return tot

def get_chain(info):
    chain_identifiers = [] 
    
    # split the input string by comma 
    try:
        pairs = info.split(",")
    #pairs = info.split(",")
    except:
        if np.isnan(info):
            return np.nan
        #print(info)
        
    for pair in pairs:
        # remove leading/trailing spaces
        pair = pair.strip() 
        # Extract the cahin identifer
        chain_id = pair[0] 
        # append id to chain identifiers list
        chain_identifiers.append(chain_id)
            
    return chain_identifiers
```
